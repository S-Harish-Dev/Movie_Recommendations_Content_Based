**CountVectorizer** (Just Counts the words) :  It converts a collection of text documents into a matrix of token counts, representing the frequency of words (or tokens) in each document.

**[*CountVectorizer*]** processes text by:

**Tokenization**: Splitting text into individual words or tokens (by default, words are split based on whitespace and punctuation is ignored).



**Vocabulary Building**: Creating a vocabulary of unique tokens (words) from the entire corpus (collection of documents).



**Counting**: Counting the frequency of each token in each document.


**Matrix Creation**: Representing the corpus as a sparse matrix where rows correspond to documents and columns correspond to tokens in the vocabulary. Each cell contains the count of a token in a document.

Works:


documents = [
    "I love to code",
    "Code is fun",
    "I love to learn"
]


----------------------------


Output:
[[1 0 0 0 1 1]  # "I love to code"

[1 1 1 0 0 0]  # "Code is fun"

[0 0 0 1 1 1]] # "I love to learn"


-------------------------------


where 
columns : I Love to Code fun learn    {Unique Text (Words like in Dictonaries)}

Rows:  Takes the presence of word from the text  {It shows 1 in the place of word in the dictonarie}

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Cv  =  CountVectorizer(max_features=(10000),stop_words='english') # removes stop words (, . @ ``) in english language and Max_Feature is Rows (uniuqe data and need all)
Cv
```

Cv.fit_transform returns a sparse matrix (in CSR format) to save memory, as most cells are zeros (most words don’t appear in most documents).

This Doesn't Work (processed = Cv.fit_transform(movies['tags'].values)):



Because the values is mxed of types making it string only helps CountVector Work

```python
processed = Cv.fit_transform(movies['tags'].values.astype('U'))
```

**astype('U')** : Non-string data (e.g., integers, floats) is converted to their string representation (e.g., 1 becomes '1', 3.14 becomes '3.14').

```python
processed = processed.toarray()
```

```python
processed.shape
```

```python
processed
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similairty = cosine_similarity(processed)
```

```python
similairty
```

We want to make Similairty as a dictonaries with index value of movie rows and inside another(0 to end , scores for each relation )

```python
name = str(input('Movie Name'))

movies_index = movies[movies['title']==name].index[0] # Gets orignal index of the movie

# index gives [Index([2], dtype='int64')] need only that (2) in the first index

int(movies_index)
```

Syntax: lambda x: x[1] is a lambda function that:


Takes an input x (each item in the iterable being sorted).


Returns x[1], the second element of x.

```python
distance  = sorted(enumerate(similairty[int(movies_index)]), reverse=True , key=lambda x:x[1])
distance
```

```python

top_5 = distance[0:5]
top_5
```

```python
# To get the names of the movie

for i in top_5:
    print(movies.iloc[i[0]].title)   # i is (2, np.float64(1.0000000000000002) and i[0] gets first index of it
```

```python
def Recommend(name):
    movies_index = movies[movies['title']==name].index[0]
    distance  = sorted(enumerate(similairty[int(movies_index)]), reverse=True , key=lambda x:x[1])
    for i in distance[0:5]:
        print(movies.iloc[i[0]].title) 
```
